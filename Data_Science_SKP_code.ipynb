```python
data = {
    'Genre': ['Action', 'Comedy', 'Drama', 'Action', 'Comedy', 'Drama'],
    'Rating': [4.2, 3.9, 4.1, 4.3, 3.8, 4.0]
}
Netflix_Dataset = pd.DataFrame(data)
```

```python
genre_ratings = Netflix_Dataset.groupby('Genre')['Rating'].mean().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Genre', y='Rating', data=genre_ratings)
plt.title('Average Rating by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
